```python
import requests
from bs4 import BeautifulSoup

res = requests.get('https://www.procyclingstats.com/race/giro-d-italia/2022/stage-9')
soup = BeautifulSoup(res.content, 'html.parser')
items = soup.select('div.content div:nth-child(5) table')
for data in items:
    print('name:', data.find('a').get_text(), 'time:', data.find('span').get_text())
```
