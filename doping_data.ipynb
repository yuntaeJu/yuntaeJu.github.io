```python
import requests
from bs4 import BeautifulSoup

res = requests.get('https://en.wikipedia.org/wiki/List_of_doping_cases_in_sport_(A)')
soup = BeautifulSoup(res.content, 'html.parser')

items = soup.select('div.mw-parser-output > table tbody ')

for item in items:
    data = item.get_text()

print(data)
```
